# Part 1 —  Setup & Personalization

```python
student_id = "0432310005101096"

# Extract R (last two digits)
R = int(student_id[-2:])

# Define new range
range_max = R / 10

# Define new radius
radius = R / 10

# Count letters only (ignore spaces)
num_letters = 19 # Md. Robiul Hassan Rabin = 19

# Number of random points
n = 200 * num_letters

print("R =", R)
print("Range = [0,", range_max, "]")
print("New Circle Radius =", radius)
print("Number of Letters =", num_letters)
print("Total Random Points =", n)

# Generate random points
x = np.random.uniform(0, range_max, n)
y = np.random.uniform(0, range_max, n)
print(x)
print(y)
```

#  Part 2 — Simulation Logic

```python
# 1. Function to check inside circle
def is_inside_circle(x, y, r):
    return (x**2 + y**2) <= r**2

# 2. Two Random Methods
half = n // 2

inside_count = 0
outside_count = 0

# First Half: random.uniform()
for _ in range(half):
    x = random.uniform(0, range_max)
    y = random.uniform(0, range_max)

    if is_inside_circle(x, y, radius):
        inside_count += 1
    else:
        outside_count += 1

# Second Half: numpy.random.rand()
for _ in range(half):
    x = np.random.rand() * range_max   # scaled to [0, R/10]
    y = np.random.rand() * range_max

    if is_inside_circle(x, y, radius):
        inside_count += 1
    else:
        outside_count += 1

# 3. Final Running Counts
print("Total Points =", n)
print("Points Inside Circle =", inside_count)
print("Points Outside Circle =", outside_count)
```

# Part 3 — Output & Analysis



```python
# Function to run simulation
def run_simulation(n):
    half = n // 2
    inside_count = 0

    # First half - random.uniform()
    for _ in range(half):
        x = random.uniform(0, range_max)
        y = random.uniform(0, range_max)
        if is_inside_circle(x, y, radius):
            inside_count += 1

    # Second half - np.random.rand()
    for _ in range(half):
        x = np.random.rand() * range_max
        y = np.random.rand() * range_max
        if is_inside_circle(x, y, radius):
            inside_count += 1

    # Adjusted π formula
    pi_est = (inside_count / n) * ((range_max**2) / (radius**2))

    error = abs(math.pi - pi_est)

    return inside_count, pi_est, error


# Print Table Header
print("Range Used | Total Points | Points Inside | Estimated Pi | Error from math.pi")
print("-"*75)

# Run 5 simulations
for i in range(1, 6):
    n = n * i
    inside, pi_value, error = run_simulation(n)

    print(f"[0,{range_max}] | {n} | {inside} | {pi_value:.6f} | {error:.6f}")
```
