# Necessary Libraries

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
import random
import math
import pandas as pd
```

# Task- 1: Conceptual Understanding

```python
# Semi-major and semi-minor axes of ellipse
a = 2
b = 1

# Rectangle limits containing the quarter ellipse
x_min, x_max = 0, 2
y_min, y_max = 0, 1

# Calculate rectangle area
rectangle_area = (x_max - x_min) * (y_max - y_min)

# True area of quarter ellipse
true_area = (math.pi * a * b) / 4

print("True Area of Quarter Ellipse =", true_area)
```

# Task 2: Simulation function

```python
def monte_carlo_simulation(N):

    # Generate N random points in the rectangle
    x = np.random.uniform(x_min, x_max, N)
    y = np.random.uniform(y_min, y_max, N)

    # Check if points satisfy ellipse equation
    inside = (x**2)/(a**2) + (y**2)/(b**2) <= 1

    # Count points inside ellipse
    points_inside = np.sum(inside)

    # Estimate area
    estimated_area = (points_inside / N) * rectangle_area

    return points_inside, estimated_area
```

# Task 3: Experiment with different N values

```python
# True area of quarter ellipse
true_area = (math.pi * a * b) / 4

N_values = [100, 1000, 10000]

results = []

for N in N_values:

    points_inside, estimated_area = monte_carlo_simulation(N)

    error = abs(true_area - estimated_area)

    results.append([N, points_inside, estimated_area, error])

# Create result table
df = pd.DataFrame(results, columns=[
    "Total Points (N)",
    "Points Inside Ellipse",
    "Estimated Area",
    "Error"
])

print("True Area of Quarter Ellipse:", true_area)
df
```

# Task 4: Analysis

**1. How does the estimate improve with larger N?**

As the number of random points N increases, the estimated area becomes closer to the true area. This happens because a larger number of samples better represents the actual distribution of points inside the rectangle.

**2. Why does Monte Carlo work for this problem?**

Monte Carlo works because it uses random sampling to estimate probability.
The fraction of points inside the ellipse approximates the probability that a random point in the rectangle lies inside the ellipse. Multiplying this probability by the rectangle area gives an estimate of the ellipse area.
