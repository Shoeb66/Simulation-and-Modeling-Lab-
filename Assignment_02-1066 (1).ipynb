# Necessary Libraries

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Creating Personal Matrix as A

```python
# ID Matrix
d1 = 6
d2 = 6

A = np.array([[d1+2, d2+1],[2*d1, d2+2]])

# Show Matrix
print("Matrix A: \n",A,"\n")
```

# 2. Analyze Matrix A

```python
# Shape of A
size_A = A.shape

# Determinant
det_A = np.linalg.det(A)

# Rank
rank_A = np.linalg.matrix_rank(A)

# Eigenvalues
eigenvalA = np.linalg.eigvals(A)

# Inverse (only if determinant ≠ 0)
if np.linalg.det(A) != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "Matrix is singular, cannot compute inverse"

# Display results
print("Shape of A:", size_A,"\n")
print("Determinant of A:", det_A,"\n")
print("Rank of A:", rank_A,"\n")
print("Eigenvalues of A:", eigenvalA,"\n")
print("Inverse of A:\n", inv_A,"\n")
```

# 3. Value Change Experiment

```python
# New Matrix B
B = np.array([[d1+2, d2+1],[2*d1, d2-2]]) # 4th element as d2-2 instead of d2+2

# Show Matrix
print("Matrix B:\n",B,"\n")
```

# 4. Re-analyze Matrix B

```python
# Shape of B
size_B = B.shape

# Determinant
det_B = np.linalg.det(B)

# Rank
rank_B = np.linalg.matrix_rank(B)

# Eigenvalues
eigenvalB = np.linalg.eigvals(B)

# Inverse (only if determinant ≠ 0)
if np.linalg.det(B) != 0:
    inv_B = np.linalg.inv(B)
else:
    inv_B = "Matrix is singular, cannot compute inverse"

# Display results
print("Shape of B:", size_B,"\n")
print("Determinant of B:", det_B,"\n")
print("Rank of B:", rank_B,"\n")
print("Eigenvalues of B:", eigenvalB,"\n")
print("Inverse of B:\n", inv_B,"\n")
```

# 5. Comparison & Explanation

**Q1. How did the determinant change and why?**  
A: For a 2×2 matrix, determinant = ad − bc.  
  
      Det(A) = (8×8) − (7×12) = 64 − 84 = −20
      Det(B) = (8×4) − (7×12) = 32 − 84 = −52

The determinant became more negative (from −20 to −52).
This happened because we reduced one diagonal value, which lowered the product of the main diagonal (ad). Since the determinant depends directly on these products, changing one value caused a noticeable difference.  
  
  
**Q2. Did the rank change? Explain.**  
A: No, the rank did not change. Both matrices still have non-zero determinants, which means both are full rank (rank = 2). Since neither determinant is zero, the rank remains the same.  
  
  
**Q3. How did the eigenvalues respond to the value change?**  
A: The eigenvalues changed because both the determinant and the trace (sum of diagonal elements) changed. When we decreased one diagonal element, the trace became smaller and the determinant changed significantly. Since eigenvalues are calculated using these values, they shifted accordingly.  


**Q4. Is B easier or harder to invert than A? Why?**  
A: B is slightly easier to invert. This is because B has a larger absolute determinant (|−52|) compared to A (|−20|). A larger determinant generally means the matrix is farther from zero and more stable when finding the inverse.
